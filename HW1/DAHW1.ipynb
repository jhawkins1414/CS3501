$P(C)$ = the prior (.5 for healthy and dementia) <br>
$C_1$ = Dementia <br>
$C_2$ = Healthy <br>
$P(X_1)$ = evidence from right brain <br>
$P(X_2)$ = evidence from left brain <br>
$P(X | C) = P(X_1 | C) * P(X_2 | C)$ <br>
$P(C | X) = (P(X | C) * P(C)) / (P(X))$


```python
import numpy as np
import math
import scipy as sp
import pandas as pd

data = df.groupby(['TrainData','Dementia'])

healthyRight = data.get_group((1,0))['RightHippoVol']
healthyLeft = data.get_group((1,0))['LeftHippoVol']
dementiaLeft = data.get_group((1,1))['LeftHippoVol']
dementiaRight = data.get_group((1,1))['RightHippoVol']

healthyRightMean = np.mean(healthyRight)
#print("Healthy Right Mean = ", healthyRightMean)
healthyLeftMean = np.mean(healthyLeft)
#print("Healthy Left Mean = ", healthyLeftMean)
dementiaRightMean = np.mean(dementiaRight)
#print("Dementia Right Mean = ", dementiaRightMean)
dementiaLeftMean = np.mean(dementiaLeft)
#print("Dementia Left Mean = ", dementiaLeftMean)

#print("\n")

healthyRightSD = np.std(healthyRight)
#print("Healthy Right SD = ", healthyRightSD)
healthyLeftSD = np.std(healthyLeft)
#print("Healthy Left SD = ", healthyLeftSD)
dementiaRightSD = np.std(dementiaRight)
#print("Dementia Right SD = ", dementiaRightSD)
dementiaLeftSD = np.std(dementiaLeft)
#print("Dementia Left SD = ", dementiaLeftSD)

untrainData = df.loc[df['TrainData'] == 0]
n = len(untrainData)
index = untrainData.index.values
correct = 0
dems = 0
heals = 0

for x in index:
    healthy = sp.stats.norm.pdf(untrainData.loc[x,'RightHippoVol'], healthyRightMean, healthyRightSD) * sp.stats.norm.pdf(untrainData.loc[x,'LeftHippoVol'], healthyLeftMean,healthyLeftSD) * .5 
    dementia = sp.stats.norm.pdf(untrainData.loc[x,'RightHippoVol'], dementiaRightMean, dementiaRightSD) * sp.stats.norm.pdf(untrainData.loc[x,'LeftHippoVol'], dementiaLeftMean,dementiaLeftSD) *.5
    probX = healthy + dementia
    likelyHealthy = healthy/probX
    likelyDementia = dementia/probX
    if likelyHealthy > likelyDementia:
        heals = heals+1
        if df.iloc[x]["Dementia"]==0:
            correct = correct+1
    else:
        dems = dems+1
        if df.iloc[x]["Dementia"]==1:
            correct = correct+1
            
acc = correct/n * 100

print("\n")
print("The model predicted", dems, "cases of dementia and", heals, "healthy cases")
print("The model predicted", correct, "cases correctly")
print("The model had a", acc, "% accuracy rate")
```
